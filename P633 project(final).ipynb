# OUTLIER'S TREATMENT

```python
# now to treat the outliers we have :-
# capping :- Replacing outlier values
# in capping all outlier values will be replaced by upper extreme or lower extreme values
def outlier_detection(data,colname):
    q1 = data[colname].quantile(0.25)   # lower quartile
    q3 = data[colname].quantile(0.75)   # upper quartile
    iqr = q3-q1   # calculating inter quartile range

    upper_extreme = q3 + (1.5*iqr)
    lower_extreme = q1 - (1.5*iqr)

    return print('lower_extreme:-',lower_extreme,', upper_extreme:-',upper_extreme,', q1:-',q1,', q3:-',q3)
```

```python
# Outlier Treatment (Capping)
def treat_outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + (1.5 * iqr)
    lower = q1 - (1.5 * iqr)
    df.loc[df[col] > upper, col] = upper
    df.loc[df[col] < lower, col] = lower
```

```python
for col in ['Open', 'High', 'Low', 'Close', 'Adj_Close']:
    treat_outliers(df, col)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler

from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
```

```python
# Load dataset
df = pd.read_csv("AAPL (5).csv", encoding="utf-8-sig")

# Clean headers
df.columns = df.columns.astype(str).str.strip()
df.columns = df.columns.str.replace(r"\s+", " ", regex=True)

# Remove duplicate columns if any
df = df.loc[:, ~df.columns.duplicated()]

print(" Columns:", df.columns.tolist())
```

```python
# Detect Date Column
date_col = None
for col in df.columns:
    c = col.lower().replace(" ", "").replace("_", "")
    if c in ["date", "datetime", "timestamp", "time"]:
        date_col = col
        break

# If not found, first column might be Date
if date_col is None:
    if "unnamed" in df.columns[0].lower() or df.columns[0].lower() in ["index"]:
        date_col = df.columns[0]

if date_col is None:
    raise ValueError(" Date column not found in your CSV file.")

print(" Date Column Found:", date_col)

# Detect Target Column
target_col = None
for col in df.columns:
    c = col.lower().replace(" ", "").replace("_", "")
    if c in ["adjclose", "adj_close", "adjustedclose", "adjustedcloseprice"]:
        target_col = col
        break

if target_col is None:
    for col in df.columns:
        if col.lower().strip() == "close":
            target_col = col
            break

if target_col is None:
    raise ValueError(" Target column not found (Adj_Close / Close).")

print(" Target Column Found:", target_col)
```

```python
# Rename and convert Date
df.rename(columns={date_col: "Date"}, inplace=True)
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df = df.dropna(subset=["Date"])
df = df.sort_values("Date")
df.set_index("Date", inplace=True)

# Fix target column values if string currency/comma exists
df[target_col] = df[target_col].astype(str).str.replace(",", "").str.replace("$", "")
df[target_col] = pd.to_numeric(df[target_col], errors="coerce")

# Drop missing target values
df = df.dropna(subset=[target_col])

# Convert to Business frequency and forward fill
df = df.asfreq("B")
df[target_col] = df[target_col].ffill()

# Final series
ts = df[target_col].dropna()

print(" Time Series Length:", len(ts))
ts.head()
```

```python
split_ratio = 0.8
split_idx = int(len(ts) * split_ratio)

train = ts.iloc[:split_idx]
test  = ts.iloc[split_idx:]

print("Train size:", len(train))
print("Test size :", len(test))

plt.figure(figsize=(12,5))
plt.plot(train.index, train, label="Train")
plt.plot(test.index, test, label="Test")
plt.title("Train-Test Split")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def mape(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

def evaluate_model(y_true, y_pred, model_name):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae  = mean_absolute_error(y_true, y_pred)
    mp   = mape(y_true, y_pred)
    return {"Model": model_name, "RMSE": rmse, "MAE": mae, "MAPE": mp}
```

```python
arima_model = ARIMA(train, order=(5,1,0))
arima_fit = arima_model.fit()

arima_pred = arima_fit.forecast(steps=len(test))

arima_results = evaluate_model(test, arima_pred, "ARIMA")
arima_results
```

```python
plt.figure(figsize=(12,5))
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, arima_pred, label="ARIMA Forecast")
plt.title("ARIMA Forecast vs Actual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
sarima_model = SARIMAX(train, order=(1,1,1), seasonal_order=(1,1,1,12))
sarima_fit = sarima_model.fit(disp=False)

sarima_pred = sarima_fit.forecast(steps=len(test))

sarima_results = evaluate_model(test, sarima_pred, "SARIMA")
sarima_results
```

```python
plt.figure(figsize=(12,5))
plt.plot(test.index, test, label="Actual")
plt.plot(test.index, sarima_pred, label="SARIMA Forecast")
plt.title("SARIMA Forecast vs Actual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
if len(ts) < 200:
    print(" Warning: Dataset is small for LSTM. Still trying...")

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(ts.values.reshape(-1, 1))

def create_sequences(data, seq_len=60):
    X, y = [], []
    for i in range(seq_len, len(data)):
        X.append(data[i-seq_len:i, 0])
        y.append(data[i, 0])
    return np.array(X), np.array(y)

seq_len = 60
if len(scaled_data) <= seq_len:
    seq_len = max(5, len(scaled_data)//2)

X, y = create_sequences(scaled_data, seq_len)

if len(X) == 0:
    raise ValueError(f" Not enough data for LSTM sequences. Need > {seq_len}")

X = X.reshape(X.shape[0], X.shape[1], 1)

split_index = int(len(X) * 0.8)
X_train, X_test = X[:split_index], X[split_index:]
y_train, y_test = y[:split_index], y[split_index:]

print(" LSTM Shapes:", X_train.shape, X_test.shape)
```

```python
lstm_model = Sequential([
    LSTM(64, return_sequences=True, input_shape=(seq_len, 1)),
    Dropout(0.2),
    LSTM(64),
    Dropout(0.2),
    Dense(1)
])

lstm_model.compile(optimizer="adam", loss="mse")

history = lstm_model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)
```

```python
lstm_pred_scaled = lstm_model.predict(X_test)
lstm_pred = scaler.inverse_transform(lstm_pred_scaled)

y_test_actual = scaler.inverse_transform(y_test.reshape(-1,1))

lstm_results = evaluate_model(y_test_actual, lstm_pred, "LSTM")
lstm_results
```

```python
test_dates = ts.index[seq_len + split_index:]

plt.figure(figsize=(12,5))
plt.plot(test_dates, y_test_actual, label="Actual")
plt.plot(test_dates, lstm_pred, label="LSTM Forecast")
plt.title("LSTM Forecast vs Actual")
plt.legend()
plt.grid(True)
plt.show()
```

```python
future_days = 30
future_dates = pd.date_range(start=ts.index[-1] + pd.Timedelta(days=1),
                             periods=future_days, freq="B")

# ARIMA Future
arima_future = arima_fit.forecast(steps=future_days)
arima_forecast_df = pd.DataFrame({"Date": future_dates, "ARIMA_Forecast": arima_future.values})

# SARIMA Future
sarima_future = sarima_fit.forecast(steps=future_days)
sarima_forecast_df = pd.DataFrame({"Date": future_dates, "SARIMA_Forecast": sarima_future.values})

# LSTM Future
last_seq = scaled_data[-seq_len:].reshape(1, seq_len, 1)
lstm_future = []

for _ in range(future_days):
    pred = lstm_model.predict(last_seq, verbose=0)[0][0]
    lstm_future.append(pred)
    last_seq = np.append(last_seq[:, 1:, :], [[[pred]]], axis=1)

lstm_future_prices = scaler.inverse_transform(np.array(lstm_future).reshape(-1,1))
lstm_forecast_df = pd.DataFrame({"Date": future_dates, "LSTM_Forecast": lstm_future_prices.flatten()})

# Merge all forecasts
final_forecast = arima_forecast_df.merge(sarima_forecast_df, on="Date").merge(lstm_forecast_df, on="Date")
final_forecast.head()
```
